```python
import pandas as pd

df = pd.read_csv('./face_recognition/examples/csvdata.csv')
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot(df['time(s)'], df['CPU Temperature(\'C)'])
plt.xlabel('time(s)')
plt.ylabel('CPU Temperature(\'C)')
plt.xticks(np.arange(0, 5000, 1000))
plt.yticks(np.arange(35, 55, 5))
plt.axhline(y=50.8, xmin=0, xmax=5000, color='g')
plt.text(0, 50.8, "50.8")
plt.show()
```

```python
plt.plot(df['time(s)'], df['FPS(1/s)'])
plt.xlabel('time(s)')
plt.ylabel('FPS(1/s)')
plt.xticks(np.arange(0, 5000, 1000))
plt.yticks(np.arange(0, 5, 1))
plt.show()
```
